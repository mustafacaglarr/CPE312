```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x = np.arange(0,10,0.1)

y = np.sin(x)

plt.plot(x,y)

plt.title("sinus grafigi")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(150,500,50)

plt.scatter(x,y,c=colors,s=sizes)

plt.title("Nokta grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

```python
data = np.random.randn(500)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.xlabel("değerler")
plt.ylabel("Frequency")
plt.show()
```

```python
x = ['A','B','C','D','E']
y = np.random.randint(1,10,5)

plt.bar(x,y)
plt.title("sinüs grafigi")
plt.xlabel("katagoriler")
plt.ylabel("değerler")

plt.show()
```
