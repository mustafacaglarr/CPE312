```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x = np.arange(0,10,0.1)

y = np.sin(x)

plt.plot(x,y)

plt.title("sinus grafigi")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.show()
```
